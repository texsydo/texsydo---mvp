```kotlin
import arrow.core.Either
import arrow.core.Either.*
import java.nio.file.Files
import arrow.core.flatMap
import software.math.tsd.webmvp.dorep.jekyll.newTempJekyllDoRepInstaller
import kotlin.io.path.Path
import software.math.tsd.webmvp.dorep.jekyll.build
import software.math.tsd.webmvp.dorep.jekyll.install
```

# DoRep for Jekyll

```kotlin
val src = Path("blogs").toAbsolutePath()
val dst = Files.createTempDirectory("dorep-for-jekyll---web---tsd-web---mvp")
```

## Installation

```kotlin
val cli = newTempJekyllDoRepInstaller()
    .flatMap { it.install() }
    .onLeft { println(it) }
    .onRight { println(it) }
    .getOrNull()

"DoRep for Jekyll installed at file:///${cli?.binPath}"
```

## Running

```kotlin
"Building article file:///$src to file:///$dst"
```

```kotlin
Right(cli)
    .flatMap { it.build(src, dst) }
    .onLeft { println(it) }

"Article built at file:///$dst"
```

Now you can serve `dst/_site` as a static website. The sample article will be
 at the path `/txt-article`.
